```python
import torch
import matplotlib.pyplot as plt

image_path = "fridge 2.jpg"  # Make sure this matches your uploaded image filename

# Load YOLOv5 pretrained model from the local yolov5 repo
model = torch.hub.load('.', 'yolov5s', source='local')

# Run inference
results = model(image_path)

# Print and show results (image with bounding boxes)
results.print()
results.show()

# Extract detected labels
labels = results.xyxyn[0][:, -1].cpu().numpy()
label_names = results.names
detected_items = [label_names[int(label)] for label in labels]
detected_items = list(set(detected_items))  # Remove duplicates

print("\n📦 Detected Ingredients:")
for item in detected_items:
    print("-", item)
```

```python
# Prompt user for any missing or additional ingredients
user_input = input("\n📝 Enter any additional or missing ingredients (comma-separated): ")

# Split the input string into a list and clean up whitespace
additional_items = [item.strip().lower() for item in user_input.split(',') if item.strip()]

# Combine detected and additional ingredients
final_ingredients = detected_items + additional_items
final_ingredients = list(set(final_ingredients))  # Remove duplicates again

print("\n🥗 Final Ingredient List:")
for ingredient in final_ingredients:
    print("-", ingredient)
```
